Смотрим тип данных в столбце с номером образца из таблицы со списком образцов. Тип object не пойдет, его следует заменить на числовой int.

```python
animals = pd.read_excel('C:/Users/79269/Desktop/Python/Тестовые/Ksitest/Полный список животных.xlsx')
animals
```

```python
#сразу удаляем дубли в столбце с инвентарным номером, чтобы не было путаницы при объединении
animals['Инвентарный номер'] = animals['Инвентарный номер'].drop_duplicates()
```

```python
#в столбце номер образца нужно убрать приставку KSI_ у номеров образцов, чтобы это не мешало переводить данные в число
samples['Номер образца'] = samples['Номер образца'].replace('KSI_','', regex=True)
```

```python
samples
```

```python
#переводим тип данных в столбце номер образца в числовой
samples['Номер образца']=pd.to_numeric(samples['Номер образца'], errors='coerce')
```

```python
samples['Номер образца'].dtypes # проверяем, что все ок
```

```python
animals['Инвентарный номер'].dtypes
```

```python
#переименовываем стобец номер образца в инвентарный номер как в таблице animals для объединения таблиц samples и animals по столбцу Инвентарный номер
samples.columns=['Инвентарный номер']
print(samples['Инвентарный номер'].dtypes) #убеждаемся, что с типом данных все ок
```

Объединяем таблицы , используя left merge:

```python
samples_animals= samples.merge(animals, how='left', on='Инвентарный номер')
samples_animals
```

```python
#проверяем, что с новой таблице все значения в столбце Инвентарный номер уникальны
samples_animals['Инвентарный номер'].value_counts()
```

```python
#на всякий случай убираем все пропуски в столбце с Идентификатором, так как строки с пропусками в идентификаторе и остальных столбцах не несут смысла для анализа
#в результате видим, что получилось 30 строк, при этом их кол-во совпадает с кол-вом заполненных дат рождения в пустой таблице для заполнения
samples_animals= samples_animals.dropna(subset=['Идентификатор'])
samples_animals
```

В целом в конце получилась таблица, которая почти совпадает по валидационному столбцу из исходного пустого файла по дате рождения особей. я нашла неск различий в датах(пометила желтым), где-то даты не совпадали или были просто пропущенные строчки, скорее всего это связано с тем, что я удаляла дубликаты и пустые строки у идентификаторов для большей чистоты данных.

Помимо этого не нашла в смерженной таблице значения № of bundle, оставила в результате его пустым. и также значение test required. Возможно, имеется в виду наличие мутаций генов, вследствие которых отбраковываются особи из селекции.
